Explicaci칩n m치s detallada de los aut칩matas celulares usados:
* [Regla 30](https://en.wikipedia.org/wiki/Rule_30)

```python
dim = 300
universe = np.zeros([dim,dim], dtype=int)
#universe[0,:] = np.random.randint(2, size = dim) #Usar con regla 184 y 110
universe[0,dim//2] = 1                           #Usar con regla 30 y 110
for line in range(1,dim):
    for cell in range(1,dim-1):
        parents = (4 * universe[line-1, cell - 1] +
                   2 * universe[line-1, cell] +
                   1 * universe[line-1, cell + 1])
        if parents in (1,2,3,4):   #Regla 30 : caos desde reglas simpl칠rrimas, morfog칠nesis gen칠tica
#        if parents in (1,2,3,5,6): #Regla 110 : Demostrada Turing-completa (WTFFF!!!!)
#        if parents in (3,4,5,7):   #Regla 184: Simula tr치fico, part칤culas y otras cosas (?!?!?)
            universe[line, cell] = 1

plt.figure(figsize=(10,10))
plt.imshow(universe, cmap='Greys')
```
